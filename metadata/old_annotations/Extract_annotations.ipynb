```python
import pandas as pd
import re

# 1) Provide your list of Sample_IDs manually (copy from boneatlas.obs['Sanger_ID'].unique())
sample_ids = boneatlas["Sanger_ID"].unique()

# 2) Compile the pattern
sample_pattern = re.compile(r'(' + '|'.join(map(re.escape, sample_ids)) + r')')

# 3) Define extraction function
def split_cell_id(name):
    # extract sample
    m = sample_pattern.search(name)
    sample = m.group(1) if m else None
    # extract barcode (longest ACGT run)
    seqs = re.findall(r'[ACGT]{8,}', name)
    barcode = max(seqs, key=len) if seqs else None
    return pd.Series({'Sample_ID_extracted': sample, 'barcode_raw': barcode})

# 4) Apply to index
new_cols = fbm['barcode'].apply(split_cell_id)

# 5) Join back
fbm_new = fbm.join(new_cols)
```

```python
import pandas as pd

# 1) assume you already have:
#    - boneatlas: a DataFrame with boneatlas['Cell_ID']
#    - panfetal_ext_femur, fbm_ext, spine_ext, skull_ext, limb_ext

# 2) restrict every external df to only cells actually in boneatlas
valid_ids = set(boneatlas['Cell_ID'].astype(str))

def filt(df):
    return df[df['Cell_ID'].astype(str).isin(valid_ids)].copy()

pan  = filt(panfetal_ext_femur)
fbm  = filt(fbm_ext)
sp   = filt(spine_ext)
sk   = filt(skull_ext)
li   = filt(limb_ext)

# 3) Handle the panfetal↔fbm overlap:
#    * primary 'published_anno' comes from fbm
#    * any panfetal annotation for those same cells goes into 'published_anno_2'
fbm = fbm.copy()
fbm['published_anno_2'] = pd.NA

# map Cell_ID → panfetal published_anno
pan_map = pan.set_index('Cell_ID')['published_anno']

# find which fbm rows have a panfetal counterpart
mask = fbm['Cell_ID'].isin(pan_map.index)
fbm.loc[mask, 'published_anno_2'] = fbm.loc[mask, 'Cell_ID'].map(pan_map)

# 4) For panfetal-only cells (not in fbm), we want them too,
#    but with published_anno in the primary col and no secondary.
pan_only = pan[~pan['Cell_ID'].isin(fbm['Cell_ID'])].copy()
pan_only['published_anno_2'] = pd.NA

# 5) The other sources just get a blank published_anno_2
sp['published_anno_2'] = pd.NA
sk['published_anno_2'] = pd.NA
li['published_anno_2'] = pd.NA

# 6) Concatenate in this order to ensure fbm takes precedence on duplicates
combined = pd.concat([fbm, pan_only, sp, sk, li], ignore_index=True)

# 7) Make sure it's unique by Cell_ID (should already be, by construction)
combined = combined.drop_duplicates(subset='Cell_ID')

# 8) Reorder to exactly five columns
combined = combined[['Cell_ID', 'Sanger_ID', 'published_anno', 'Organ', 'published_anno_2']]

# 9) Quick sanity check
assert combined.shape[1] == 5
print("Final shape:", combined.shape)   # should be (602010, 5)
```
