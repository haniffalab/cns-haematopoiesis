## Make sure sample IDs in anndata and metadata (Run_ID) coresspond and keep intersecting samples 

```python
# List of values to exclude because these samples were multiplexed and map to multiple rows in the anndata and metadata columns
# Will process these samples separately using demultiplexed barcodes 

values_to_exclude = [
    'HCA_BN_F13109710_and_HCA_BN_F13101380',
'HCA_BN_F13109710_and_HCA_BN_F13101381',
'HCA_BN_F13109711_and_HCA_BN_F13101381',
'HCA_BN_F13109711_and_HCA_BN_F13101381',
'HCA_BN_F13109712_and_HCA_BN_F13101382',
'HCA_BN_F13109712_and_HCA_BN_F13101383',
'HCA_BN_F13109713_and_HCA_BN_F13101383',
'HCA_BN_F13109713_and_HCA_BN_F13101384']
# Create a boolean mask to identify rows to exclude
mask = ~metadata['Run_ID'].isin(values_to_exclude)

# Filter the metadata DataFrame to keep only rows that do not match the excluded values
filtered_metadata = metadata[mask]
```

To merge your metadata DataFrame with your AnnData object using a common sample ID column (Run_ID in metadata and sample_id in AnnData) without changing the number of cells, you can use the Pandas merge function with the how='left' option to perform a left join. This will keep all the cells in your AnnData and add metadata for the matching sample IDs. 

```python
#Merge the metadata with the AnnData using a left join
merged_data = filtered_adata.obs.merge(filtered_metadata, how='left', left_on='sample_id', right_on='Run_ID')
```

## Add cell_ID: sample_ID followed by barcode 

```python
def create_cell_id(row):
    if row['Organ'] == 'Spine':
        # Special handling for 'Spine' Organ
        # Remove 'cellranger600_count_' and 'GRCh38-2020-A'
        run_id = row['Run_ID'].replace('cellranger600_count_', '').replace('_GRCh38-2020-A', '')
    else:
        run_id = row['Run_ID']
    
    # Concatenate 'Run_ID' and 'barcode'
    cell_id = f"{run_id}_{row['barcode']}"
    return cell_id

# Apply the create_cell_id function to create the 'Cell_ID' column
filtered_adata.obs['Cell_ID'] = filtered_adata.obs.apply(create_cell_id, axis=1)

# Now, the 'Cell_ID' column is added to your filtered_adata.obs DataFrame
```

```python
filtered_adata.obs_names=filtered_adata.obs["Cell_ID"]
```
